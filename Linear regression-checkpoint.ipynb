```python
#Importing required libraries
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd 
import pylab as pl
import seaborn as sns
from sklearn import preprocessing, svm
%matplotlib inline
```

```python
#Loading the dataset
data = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
data.head()
```

```python
#Statistical overview of the dataset
data.describe()
```

```python
#Line graph of the two individual functions namely 'Hours' and 'Scores'
data.plot(figsize=(20,6))
```

```python
#Histogram plot of the two individual functions namely 'Hours' and 'Scores'
data.hist()
```

```python
#Scatter plot of the dataset
plt.scatter(data['Hours'], data['Scores'])
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.title('Relationship between Hours and Scores', size=24)
plt.show()
```

```python
#Regression plot using seaborn
sns.lmplot(x ="Hours", y ="Scores", data=data , order = 2, ci = None)
plt.title('Hours vs Percentage', size=24)
```

```python
X = np.array(data['Hours']).reshape(-1, 1) 
y = np.array(data['Scores']).reshape(-1, 1) 
```

```python
# Separating the data into independent and dependent variables 
# Converting each dataframe into a numpy array  
# since each dataframe contains only one column 
data.dropna(inplace = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
# create linear regression object
# Splitting the data into training and testing data 
regr = LinearRegression() 
# fit linear regression  
regr.fit(X_train, y_train) 
print(regr.score(X_test, y_test)) 
```

```python
y_pred = regr.predict(X_test) 
plt.scatter(X_test, y_test, color ='r') 
plt.plot(X_test, y_pred, color ='y') 
  
plt.show() 
# Data scatter of predicted values 
```

```python
# To get the slope and intercept of the line best fit
print(regr.intercept_)
print(regr.coef_)
```

```python
# predictions using Linear regression equation
y_pred=regr.intercept_+regr.coef_*9.25
print('Predicted score if a student studies for 9.25 hours/day',y_pred)
```

```python
# predictions using scikit learn
print(regr.predict([[9.25]]))
```

```python
#Evaluating the Algorithm
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
